#1. Do the height and weight columns have the appropriate data types?


```python
import pandas as pd

# Assuming DataFrame is named 'df' and the columns are 'Height' and 'Weight'
df['Height'] = df['Height'].str.extract('(\d+)').astype(float)
df['Weight'] = df['Weight'].str.extract('(\d+)').astype(float)


df.info()
```

#2. Can you separate the joined column into year, month and day column?

```python
print(df['Joined'])
```

```python
# Assuming DataFrame is named 'df' and the column is 'Joined'
df['Joined'] = pd.to_datetime(df['Joined'])
df['Year'] = df['Joined'].dt.year
df['Month'] = df['Joined'].dt.strftime('%b')
df['Day'] = df['Joined'].dt.day

df
```

#3. Can you clean and transform the value, wage and release clause columns into columns of float?

```python
df['Value'], df['Wage'], df['Release Clause']
```

```python
# Convert columns to string data type
df['Value'] = df['Value'].astype(str)
df['Wage'] = df['Wage'].astype(str)
df['Release Clause'] = df['Release Clause'].astype(str)
# Remove unwanted characters and convert to float
df['Value'] = df['Value'].str.replace('[€]', '', regex=True).astype(float)
df['Wage'] = df['Wage'].str.replace('[€]', '', regex=True).astype(float)
df['Release Clause'] = df['Release Clause'].str.replace('[€]', '', regex=True).astype(float)

def clean_and_transform_value(value):
    if value.endswith('M'):
        return int(float(value[:-1]) * 1000000)
    elif value.endswith('K'):
        return int(float(value[:-1]) * 1000)
    else:
        return int(value)

# Example usage:
value_cell = '€103.5M'
cleaned_value = clean_and_transform_value(value_cell)
print(cleaned_value)  # Output: 103500000

wage_cell = '€560K'
cleaned_wage = clean_and_transform_value(wage_cell)
print(cleaned_wage)  # Output: 560000
```

```python
# Convert columns to string data type
df['Value'] = df['Value'].astype(str)
df['Wage'] = df['Wage'].astype(str)
df['Release Clause'] = df['Release Clause'].astype(str)
# Remove unwanted characters and convert to float
df['Value'] = df['Value'].str.replace('[€, MK]', '', regex=True).astype(float)
df['Wage'] = df['Wage'].str.replace('[€, K]', '', regex=True).astype(float)
df['Release Clause'] = df['Release Clause'].str.replace('[€,M,K]', '', regex=True).astype(float)

# Select the desired columns
selected_columns = df[['Value', 'Wage', 'Release Clause']]
selected_columns
```

#4.Remove the unnecessary newline characters from all columns that have them.

```python
# Remove newline characters from all columns
df = df.replace(r'\n', '', regex=True)
```

#5. Based on the 'Joined' column, check which players have been playing at a club for more than 10 years?

```python
# Calculate the number of years a player has been at the club
df['Years_at_Club'] = pd.to_datetime('today') - df['Joined']
df['Years_at_Club'] = round((df['Years_at_Club'].dt.days / 365.25),2)

# Filter players who have been at the club for more than 10 years
players_over_10_years = df[df['Years_at_Club'] > 10][['Name', 'Years_at_Club']]

# Display the players and their joining year
print(players_over_10_years)
```

# 6. Which players are highly valuable but still underpaid (on low wages)? (hint: scatter plot between wage and value)


```python
import matplotlib.pyplot as plt
import numpy as np

# Create a scatter plot
plt.scatter(df['Wage'], df['Value'])
plt.xlabel('Wage')
plt.ylabel('Value')
plt.title('Wage vs. Value')

# Add a trend line
z = np.polyfit(df['Wage'], df['Value'], 1)
p = np.poly1d(z)
plt.plot(df['Wage'], p(df['Wage']), color='red')

# Identify highly valuable but underpaid players
threshold_value = 1000000  # Define the threshold value for "highly valuable"
threshold_wage = 100000  # Define the threshold value for "low wage"
underpaid_players = df[(df['Value'] > threshold_value) & (df['Wage'] < threshold_wage)]

# Add markers for underpaid players
plt.scatter(underpaid_players['Wage'], underpaid_players['Value'], color='green', marker='x')

# Display the plot
plt.show()
```
